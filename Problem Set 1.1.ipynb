```python
X = df[['educ', 'exper', 'tenure', 'nonwhite', 'female', 'married','numdep', 'smsa', 'northcen', 'south', 'west', 'construc', 'ndurman','trcommpu', 'trade', 'services', 'profserv', 'profocc', 'clerocc','servocc']]
X = stm.add_constant(X) 
y = df['wage']

model = stm.OLS(y, X).fit()
print(model.summary())
```

```python
The OLS regression model was used to predict the wage based on 20 independent variables. 
The R-squared value was 0.47, indicating that 47% of the variation in the wage could be explained by the independent variables. 
The model showed that education, tenure, being female, working in the trade or services industry, and having a professional occupation had a significant positive impact on the wage. 
```
